```python
import pandas as pd       # pandas 패키지를 호출하여, pd로 지정 

station_data=pd.read_csv('서울시 소방서관할 위치정보 (좌표계_ WGS1984).csv',
                 encoding='cp949')  
station_data.head()
```

```python
location_data=station_data[['위도','경도']].values[:len(station_data)].tolist()
```

```python
center = [37.541, 126.986]    # 서울시 중심 위/경도 
m=folium.Map(location=center, zoom_start=10)   # folium 지도 불러 오기 

for i in range(len(station_data)):  # len함수로 전체 대여소 갯수 파악, for문을 통해 반복 수행 
    station_id=station_data['순번'][i]  # 마우스 올렸을 때 id 표시를 위해서 id값 추출 
    max_bc_count=int(station_data['방면코드'][i])  # 거치대 수를 max_bc_count 변수 지정 
    if station_data['방면코드'][i] == 1 :  # if문을 활용하여 station type 색 지정,
        station_color='blue'
    elif station_data['방면코드'][i] == 2 : 
        station_color='red'
    else :station_color='black'
    folium.Circle(location=location_data[i],radius=max_bc_count,color=station_color,fill=True,fill_opacity=1,tooltip=station_id,popup=max_bc_count).add_to(m)
      # 채원진 원을 표시하기 위해서, fill_opacity=1로 지정, popup값은 거치대 숫자로 표시    
m
```

```python
from folium import plugins     # MarkerCluster를 사용하기 위해서 plugin기능을 호출함.


center = [37.541, 126.986]
m=folium.Map(location=center, zoom_start=10)

for i in range(len(station_data)):
    station_id=station_data['순번'][i]
    max_bc_count=int(station_data['방면코드'][i])
    if station_data['방면코드'][i] == 1 : 
        station_color='blue'
    elif station_data['방면코드'][i] == 2 : 
        station_color='red'
    elif station_data['방면코드'][i] == 3 : 
        station_color='black'   
    folium.Circle(location=location_data[i],radius=max_bc_count,color=station_color,fill=True,fill_opacity=1,tooltip=station_id,popup=max_bc_count).add_to(m)

plugins.MarkerCluster(location_data).add_to(m)  # list로 된 위치 정보를 MarkerCluster함수에서호출

m
```
